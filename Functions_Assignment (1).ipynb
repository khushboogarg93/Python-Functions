### 2. Explain the concept of function arguments and parameters in Python.

        1. *args - for a variable number of positional arguments, uses forloop generally and return as tuple if multiple :

           ex- def sum_up(*args):
                  return sum(args)

              sum_up(1, 2, 3, 4)     # returns 10
              sum_up(1, 2, 3, 4, 1)  # returns 11

        2. **kwargs - for a variable number of keyword arguments, uses dictionary iteration and return as dict:

           ex- def print_details(**kwargs):
                  for key, value in kwargs.items():
                      print(f"{key}: {value}")

              print_details(name="Khushboo", age=24)

### 5. What are iterators in Python and how do they differ from iterables?

> Iterable

  - Iterable is any object which can loop over using a for loop ex- string, list, dict, tuple, sets etc.

  - Python understand the object behind the scene and loop over.
  
  - It has a built-in method __iter__() that returns an iterator.

  - ex -
       my_list = [1, 2, 3]

       for num in my_list:  # This works because lists are iterable
       print(num)

> Iterator

  - Iterator is an object that produces values one at a time using the __next__() method.
  
  - It remembers its state, so it can continue where it left off.

  - To be an iterator, an object must implement both: __iter__() and __next__()

  - ex-
      my_list = [1, 2, 3]
      my_iterator = iter(my_list)

      print(next(my_iterator))  # Output: 1
      print(next(my_iterator))  # Output: 2
      print(next(my_iterator))  # Output: 3
      print(next(my_iterator))  # Output: would raise StopIteration as list as only 3 elements

### 6. Explain the concept of generators in Python and how they are defined.

> Generator

  - When we dont want to return result in 1 go and exhaust the memory a lot, we used a concept of Generator.

  - Generator is a special type of iterator that is defined using a function with the "yield" keyword.

  - It pauses execution at each yield and resumes from that point the next time it's called.

  - It automatically keeps track of its internal state.

  - It remembers all local variables and the exact place it was paused.

  - Regular functions uses return , generator function uses yield.

  - It takes less memory compare to regular function.

  - ex -
   def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

   counter = count_up_to(3)
   print(next(counter))  # Output: 1
   print(next(counter))  # Output: 2
   print(next(counter))  # Output: 3
   print(next(counter))  # Output: would raise StopIteration as number is 3

### 7. What are the advantages of using generators over regular functions?
- Generators have many advantages over regular functions, especially when working with large data or streams of data.

  1. Memory Efficiency - Generators yield one item at a time instead of generating the entire sequence and storing it in memory.

  ex- # Regular function
    def big_list():              
       return [i for i in range(1000000)] #it builds a full list in memory.

  ex- # Generator function
    def big_generator():         
        for i in range(1000000):
             yield i       #produces one item at a time — much lighter on RAM.


  2. Lazy Evaluation - Generators compute values only when needed, which is generally ideal as well for large or infinite sequences.

  3. Improved Performance - Since generators avoid creating and storing large data structures, they can offer better performance in terms of speed and responsiveness, especially in pipelines or stream processing.

  4. State Preservation - Generators automatically remember their execution state between yield calls — no need for manually tracking indexes or positions.

   ex-

   def step_by_step():
    yield "Step 1"
    yield "Step 2"
    yield "Step 3"

  5. Cleaner Code for Iteration - with the generator function code will be more cleaner compare to regular functions.

### 9. Explain the purpose and usage of the `map()` function in Python.

> Map() Function

  - map() function in Python is a built-in tool that allows to apply a function to every item in an iterable (like a list, tuple, etc.) and return the results as a new iterator.

  - i.e. to transform items in a collection by applying a function to each one, without writing a loop.

  - Syntax - map(function_name, iterable) # here function_name is the function name and iterable is like list, tuple, sting etc.

  - Lambda function also be used in place of function_name.

  - If the function takes more than one argument, we can pass multiple iterables as well in map

  - map() returns a map object which is an iterator, it usually wrap it with list() to see the results.

  - When we have already have a function and want to apply it quickly.

  - When working with large datasets because it's lazy/evaluated on-demand.

  - ex-
    1. Square list of number
       ex- lst = [1,2,3,4]
           result = [map(lambda x: x * x, lst)]
           print(result) #op= [1,4,9,16]

    2. Convert string to uppercase
       ex- my_lst = ["khushboo", "garg"]
           result = [map(str.upper, my_str)]
           print(result) #op= ['KHUSHBOO', 'GARG']

    3. Multi iterable
        ex- lst1 = [1,2,3,4]
            lst2 = [5,6,7,8]

  - map() returns a map object which is an iterator, it usually wrap it with list() to see the results.

> Reduce Function

  - Its not a built in function like map and others. In order to use this we have to import it from library "functools" modules

  - It is used to reduce a sequence of items to a single cumulative value by applying a binary function repeatedly i.e. To perform cumulative computation like sum, product, maximum, or string concatenation — across a sequence.

  - Syntax =
     from functools import reduce
     reduce(function, iterable[, initializer])

  - Here function is a function name calling it exactly takes 2 arguments, if we put more than that it will start giving exception.

  - iterable - sequence to process i.e. list, string, tuple etc.

  - initializer is optional, i.e. starting value for the accumulation

  - Use it when Accumulating values like sum, product, min/max, etc. and Data aggregation

  - Avoid to use reduce when logic is too complex.

  - ex-
    1. Sum a list of numbers
       ex-
       from functools import reduce
       
       lst = [1,2,3,4]
       result = reduce(lambda x, y : x+y, lst)
       print(result) #op - 10

    2. Find the longest word
       ex-
       from functools import reduce
       lst = ["Khushboo", "Garg", "Shrivastava"]

       result = reduce(lambda x,y : x if (x > y) else y, lst)
       print(result) #op- 'Shrivastava'

    3. Using initializer
       ex-
       from functools import reduce
       lst = [1,2,3]
       result = reduce(lambda x, y : x+y , lst, 4)  # here it starts with 4
       print(result) #op- 10

> Filter Function

  - filter() function in Python is a built-in tool that allows and used to filter elements from an iterable based on a function that returns True or False. i.e. To remove unwanted elements from a sequence without writing a loop.

  - It helps us to keep only the items that satisfy a certain condition.

  - Syntax - filter(function, iterable), here function is the function name calling (function must satisty true or false) and iterable is a squence process like list, tuple etc.

  - ex-
    1. Filter even numbers
    ex- nums = [1,2,3,4]
        result = [filter(lambda x : x%2 == 0, nums)]
        print(result) #op- [2,4]

    2. Find out empty strings
    ex- words = ["python", "", "code", "", "Data"]
        result = [filter(None, words)]
        print(result) #op - ['python', 'code', 'Data']

These are the basic difference between them.

```python
'''
1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list.
'''
def sum_even(lst):
  sum_nums = 0
  for num in lst:
    if num % 2 == 0:
      sum_nums += num
  return sum_nums

lst = [1,2,3,4,6,7]
print(f"Sum of all even numbers :", sum_even(lst))
```

```python
'''
3. Implement a Python function that takes a list of integers and returns a new list containing the squares of
each number.
'''
def squared_num(my_lst):
  new_lst = []
  for num in my_lst:
    if isinstance(num, int):
      new_lst.append(num ** 2)
  return new_lst

my_lst = [1,2,3,4]
print(squared_num(my_lst))
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
# prime numbers are those who are divide from iteself or 1 i.e.  2,3,5,7,11,13.....
def is_prime():
    for num in range(1, 201): # iterate range of numbers for further check
        if num < 2:
            print(f"{num} is not prime")
        else:
            is_prime = True
            for i in range(2, num): # checks number from 2 to num-1, whether num is prime or not
                if num % i == 0:
                    is_prime = False
                    break

            if is_prime:
                print(f"{num} is prime")
            else:
                print(f"{num} is not prime")

is_prime()
```

```python
'''
5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.
'''
# fibonacci series - are the numbers sum of previous 2 numbers i.e. 0,1,1,2,3,5....

class FibonacciSeries:
    def __init__(self, num):
        self.num = num
        self.count = 0
        self.a = 0 #initial first value
        self.b = 1 #initial second value

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.num:
            raise StopIteration
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return value

fib = FibonacciSeries(10)
for item in fib:
    print(item, end=',')
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def power_generator(num):
  for i in range(num+1):
    yield i**2

for val in power_generator(4):
    print(val, end=', ')
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
import gdown # download files from Google Drive

file_id = '1Ir_mf1h2LvYTOhBjE_Nc8lyVClcCNOsU'
file_name = 'C.txt'

gdown.download(f'https://drive.google.com/uc?id={file_id}', file_name, quiet=False)

def read_file_line_by_line(filename):
    with open(filename, 'r') as file:
        for line in file:
            yield line.strip()  # strip() removes any trailing newline characters

for line in read_file_line_by_line(file_name):
    print(line)
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
celsius_temps = [0, 10, 20, 30, 40, 50]

def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

fahrenheit_temps = list(map(celsius_to_fahrenheit, celsius_temps))

print(fahrenheit_temps)
```
